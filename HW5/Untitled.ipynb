```python
def parser(r_pattern):
    pat = r_pattern.split(" ")
    pattern = {"name": pat[0], "operator":pat[1], "value": pat[2]}
    return pattern
```

```python
def switcher(cva,options1):
    switcher = {
        "<=": cva <=float(options1["value"]),
        ">=": cva >=float(options1["value"]),
        "<": cva < float(options1["value"]),
        ">": cva > float(options1["value"]),
        "=": cva == float(options1["value"]),
        "==": cva == float(options1["value"]),
        "!=": cva != float(options1["value"]),
        }
    return switcher

def dividePat(pattern):
    ptrns = pattern.split("AND")
    rptr = []
    for ptr in ptrns:
        rptr.append(parser(ptr))
    
    return rptr

def fourItemNB(patterns,dataset):
    
    ptrns = dividepat(patterns)
    
    options1 = ptrns[0]
    options2 = parser(patterns[1])
    options3 = parser(patterns[2])
    options4 = parser(patterns[2])
    
    colVal1 = dataset[options1["name"]].values
    colVal2 = dataset[options2["name"]].values
    colVal3 = dataset[options3["name"]].values
    colVal4 = dataset[options4["name"]].values
    ncolVal3 = (colVal3-min(colVal3))/(max(colVal3)-min(colVal3))
    ncolVal4 = (colVal4-min(colVal4))/(max(colVal4)-min(colVal4))*8

    data0 = []
    data1 = []
    count = 0
    
    for indx in range(len(colVal1)):
        cva = colVal1[indx]
        cva2 = colVal2[indx]
        cva3 = ncolVal3[indx]
        cva4 = ncolVal4[indx]
        
        if(options1["operator"] == ">"):
            if(cva > float(options1["value"])):
                data0.append([cva,cva2,cva3,cva4])
            else:
                data1.append([cva,cva2,cva3,cva4])
                
        if(options1["operator"] == "<"):
            if(cva < float(options1["value"])):
                data0.append([cva,cva2,cva3,cva4])
            else:
                data1.append([cva,cva2,cva3,cva4])
                
    run4 = [data0,data1,[float(options1["value"])],[float(options2["value"])]]
    runT4NB(run4)

def threeItemNB(patterns,dataset):
    options1 = parser(patterns[0])
    options2 = parser(patterns[1])
    options3 = parser(patterns[2])
    
    colVal1 = dataset[options1["name"]].values
    colVal2 = dataset[options2["name"]].values
    colVal3 = dataset[options3["name"]].values
    ncolVal3 = (colVal3-min(colVal3))/(max(colVal3)-min(colVal3))

    data0 = []
    data1 = []
    count = 0
    
    for indx in range(len(colVal1)):
        cva = colVal1[indx]
        cva2 = colVal2[indx]
        cva3 = ncolVal3[indx]
        
        sw1 = switcher(cva,options1)
        sw2 = switcher(cva2,options2)
        sw3 = switcher(cva3,options3)
        
        if(options1["operator"] == ">"):
            if(cva > float(options1["value"])):
                data0.append([cva,cva2,cva3])
            else:
                data1.append([cva,cva2,cva3])
                
        if(options1["operator"] == "<"):
            if(cva < float(options1["value"])):
                data0.append([cva,cva2,cva3])
            else:
                data1.append([cva,cva2,cva3])
                
    run3 = [data0,data1,[float(options1["value"])],[float(options2["value"])]]
    runT3NB(run3)

def twoItemNB(patterns,dataset):
    options1 = parser(patterns[0])
    options2 = parser(patterns[1])
    colVal1 = dataset[options1["name"]].values
    colVal2 = dataset[options2["name"]].values
    
    data0 = []
    data1 = []
    count = 0
    
    for indx in range(len(colVal1)):
        cva = colVal1[indx]
        cva2 = colVal2[indx]
        
        if(options1["operator"] == ">"):
            if(cva > float(options1["value"])):
                data0.append([cva,cva2])
            else:
                data1.append([cva,cva2])
                
        if(options1["operator"] == "<"):
            if(cva < float(options1["value"])):
                data0.append([cva,cva2])
            else:
                data1.append([cva,cva2])
                
    run2 = [data0,data1,[float(options1["value"])],[float(options2["value"])]]
    runT2NB(run2)
    
def oneItemNB(pattern,dataset):
    options = parser(pattern)
    colVal = dataset[options["name"]].values
    data0 = []
    data1 = []
    count = 0
    for cva in colVal:
        if(options["operator"] == ">"):
            if(cva > float(options["value"])):
                data0.append([count,cva])
            else:
                data1.append([count,cva])
        if(options["operator"] == "<"):
            if(cva < float(options["value"])):
                data0.append([count,cva])
            else:
                data1.append([count,cva])
        count = count +1
    
    run1 = [data0,data1,[float(options["value"])]]
    runT1NB(run1)

def checkOP(options,cva):
    isit = false
    if(options["operator"] == ">"):
            if(cva > float(options["value"])):
                isit = true
    if(options["operator"] == "<"):
        if(cva < float(options["value"])):
            isit = true
    return isit

def oneItemNM(patternM,dataset):
    opt = []
    dats = []
    
    for ptrn in patternM:
        options = parser(ptrn)
        opt.append(options)
        dats.append([])
        
    colVal = dataset[opt[0]["name"]].values
    
    count = 0
    for cva in colVal:
        
        for ind in range(len(opt)):
            o = opt[ind]
            if(checkOP(o,cva)):
                dats[ind].append([count,cva])
                continue
        
        count = count +1
    
    run1 = [data0,data1,[float(options1["value"])]]
    runT1NM(run1)
```
